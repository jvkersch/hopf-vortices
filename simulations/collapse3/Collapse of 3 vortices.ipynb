Comparison of Hopf and RK4, no regularization---------------------------------------------

```python
# Load data times_sphere, vortices, energies_sphere, moments_sphere = \        load_variables('data/collapse3_sphere_sigma00_sim.mat', ['times', 'vortices', 'energies', 'moments'])times_rk4, vortices, energies_rk4, moments_rk4 = \        load_variables('data/collapse3_rk4_sigma00_sim.mat', ['times', 'vortices', 'energies', 'moments'])times_lp, vortices, energies_lp, moments_lp = \        load_variables('data/collapse3_lp_sigma00_sim.mat', ['times', 'vortices', 'energies', 'moments'])times_mp, vortices, energies_mp, moments_mp = \        load_variables('data/collapse3_mp_sigma00_sim.mat', ['times', 'vortices', 'energies', 'moments'])
```

```python
# Compute energy/moment errorsenergy_error_sphere = energies_sphere - energies_sphere[0]energy_error_rk4 = energies_rk4 - energies_rk4[0]energy_error_lp = energies_lp - energies_lp[0]energy_error_mp = energies_mp - energies_mp[0]
```

```python
# Compute moment errorsdef moment_norm_error(moments):    n = moments.shape[0]    norm_error = np.empty(n)    for k in xrange(0, n):        norm_error[k] = LA.norm(moments[k,:] - moments[0, :])    return norm_error    
```

```python
moment_norm_error_sphere = moment_norm_error(moments_sphere)moment_norm_error_rk4 = moment_norm_error(moments_rk4)moment_norm_error_lp = moment_norm_error(moments_lp)moment_norm_error_mp = moment_norm_error(moments_mp)
```

```python
# Energy comparison figurefig = figure()ax = fig.add_subplot(111)plot(times_sphere, energy_error_sphere, 'k', label='Hopf')plot(times_rk4, energy_error_rk4, 'k--', label='RK4')plot(times_lp, energy_error_lp, 'k-.', label='LP')plot(times_mp, energy_error_mp, 'k:', label='MP')ax.yaxis.set_major_formatter(sfmt)ax.yaxis.set_major_locator(MaxNLocator(6))xlabel(r'$t$', fontsize=20)ylabel('Energy error')#legend(loc='upper left')savefig('collapse3-energy-comparison-short.pdf')show()
```

```python
# Moment comparison figurefig = figure()ax = fig.add_subplot(111)plot(times_sphere, moment_norm_error_sphere, 'k', label='Hopf')plot(times_rk4, moment_norm_error_rk4, 'k--', label='RK4')plot(times_lp, moment_norm_error_lp, 'k-.', label='LP')plot(times_mp, moment_norm_error_mp, 'k:', label='MP')ax.yaxis.set_major_formatter(sfmt)ax.yaxis.set_major_locator(MaxNLocator(6))xlabel(r'$t$', fontsize=20)ylabel('Norm of the moment error')ylim((-1e-4, 6e-4))#legend(loc='upper left')savefig('collapse3-moment-comparison-short.pdf')show()
```

```python
# Load and process energy/moment errors for various sigmaenergy_error_sphere = []momentum_error_sphere  = []for sigma, postfix in [(0., 'sigma00'), (.10, 'sigma10'), (.25, 'sigma25')]:    filename = 'data/collapse3_sphere_' + postfix + '_sim.mat'    times, E, M = load_variables(filename, ['times', 'energies', 'moments'])    energy_error_sphere.append((sigma, E-E[0]))
```
