Comparison of Hopf and RK4, no regularization---------------------------------------------

```python
# Load data times, vortices, energies_sphere, moments_sphere = \        load_variables('data/collapse3_sphere_sigma00_sim.mat', ['times', 'vortices', 'energies', 'moments'])times, vortices, energies_rk4, moments_rk4 = \        load_variables('data/collapse3_rk4_sigma00_sim.mat', ['times', 'vortices', 'energies', 'moments'])
```

```python
# Compute energy/moment errorsenergy_error_sphere = energies_sphere - energies_sphere[0]energy_error_rk4 = energies_rk4 - energies_rk4[0]
```

```python
n = len(times)moment_norm_error_sphere = np.empty(n)moment_norm_error_rk4 = np.empty(n)for k in xrange(0, n):    moment_norm_error_sphere[k] = LA.norm(moments_sphere[k,:] - moments_sphere[0, :])    moment_norm_error_rk4[k] = LA.norm(moments_rk4[k,:] - moments_rk4[0, :])    
```

```python
# Energy comparison figurefig = figure()ax = fig.add_subplot(111)plot(times, energy_error_sphere, label='Hopf')plot(times, energy_error_rk4, label='RK4', linestyle='--')ax.yaxis.set_major_formatter(sfmt)ax.yaxis.set_major_locator(MaxNLocator(6))xlabel(r'$t$', fontsize=20)ylabel('Energy error')legend(loc='upper left')savefig('collapse3-energy-comparison-short.pdf')show()
```

```python
# Moment comparison figurefig = figure()ax = fig.add_subplot(111)plot(times, moment_norm_error_sphere, label='Hopf')plot(times, moment_norm_error_rk4, label='RK4', linestyle='--')ax.yaxis.set_major_formatter(sfmt)ax.yaxis.set_major_locator(MaxNLocator(6))xlabel(r'$t$', fontsize=20)ylabel('Norm of the moment error')ylim((-1e-4, 4e-4))legend(loc='upper left')savefig('collapse3-moment-comparison-short.pdf')show()
```

Comparison of sigma parameter for Hopf integrator -------------------------------------------------Simulation parameters: $h = 0.1$, $T = 15$. Regularization: $\sigma = 0.0, 0.1, 0.25$.

```python
import numpy.linalg as LA
```

```python
# Load and process energy/moment errors for various sigmaenergy_error_sphere = []momentum_error_sphere  = []for sigma, postfix in [(0., 'sigma00'), (.10, 'sigma10'), (.25, 'sigma25')]:    filename = 'data/collapse3_sphere_' + postfix + '_sim.mat'    times, E, M = load_variables(filename, ['times', 'energies', 'moments'])    energy_error_sphere.append((sigma, E-E[0]))
```

```python
# Variational energy errorlinestyles = ['-', '--', '-.']fig = figure()ax = fig.add_subplot(111)ax.autoscale(axis='x', tight=True)for (sigma, err), style in zip(energy_error_sphere, linestyles):    plot(times, err, linestyle=style, label=r'$\sigma = %.2f$' % sigma)xlabel(r'$t$', fontsize=20)ylabel('Energy error')ax.yaxis.set_major_formatter(sfmt)ax.yaxis.set_major_locator(MaxNLocator(5))ax.xaxis.set_major_locator(MaxNLocator(5))legend(loc='best')savefig('collapse3-energy-variational-short.pdf')show()
```
