```python
# Unit length preservationdef unit_length_deviation(vortices):    n, m = vortices.shape[0:2]    dev = np.zeros(n)    tmp = np.zeros(m)    for i in xrange(0, n):        for j in xrange(0, m):            tmp[j] = LA.norm(vortices[i,j,:]) - 1.        dev[i] = np.mean(tmp)    return dev
```

```python
dev_sphere = unit_length_deviation(vortices_sphere)dev_lp = unit_length_deviation(vortices_lp)dev_mp = unit_length_deviation(vortices_mp)
```

```python
fig = figure()ax = fig.add_subplot(111)plot(times_sphere, dev_sphere, label='Hopf')plot(times_lp, dev_lp, label='Lie-Poisson', linestyle='--')plot(times_mp, dev_mp, label='Midpoint', linestyle='-.')ax.yaxis.set_major_formatter(sfmt)ax.autoscale(axis='x', tight=True)xlabel(r'$t$', fontsize=20)ylabel('Unit length error')legend(loc='best')savefig('generic-40-unitlength.pdf')show()
```
